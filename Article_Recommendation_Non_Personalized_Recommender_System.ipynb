# Answer 3: use the following formula and identify the top 10 articles based on  weighted rating as mentioned below:

W = R * v + C * m / v + m               


```python
# 1:Calculate the average rating (R) for each article.
# 2: Determine the number of ratings (v) for each article.
# 3: m:  Set the minimum number of ratings required for an article to be on the recommendation list. Let's use m = 2 in this case.
# 4: Calculate the mean rating for all the articles (C).
# 5: Apply the weighted rating formula (W = R + (v / (v + m)) * C) for each article.
# 6: Sort the articles based on their weighted ratings in descending order.
# 7: Select the top 10 articles from the sorted list.
```

## Step 1: Calculate the average rating (R) for each article

waxaan sameesaneenaa colu: average_rating:
article_id inta jeer u soo laalaabte isku dar, adigoo fiirnaayo columka rating.

example:
article_id  rating
101             1
105             2
99            4
105             1
101             1
101             2   
99             3
99             5
105             3
105             2

article_id waxaan ku heestaa: 99, 101, 105 âœˆ
si loo sameeyo average_rating: 101, 3 jeer ayuu soo laalaabte ratingkiisa isku soo dar dar.
sida: 1+1+2 = 4 /3 (inta jeer u soo laalabte) = 1.3

99: wuxuu soo lalaabte: 3 jeer
99: 4 + 3 + 5: 12 / 3(inta jeer u soo laalabte) = 4

105: wuxuu soo laalaabte: 4 jeer
2 + 1 + 3 + 2 = 8 / 4(inta jeer u soo laalabte) = 2

tableka sidaan ayuu noqonaa:
article_id     average_rating
101                 1.3
105                 2
99                  4



```python
# Step 1: Calculate the average rating (R) for each article
average_ratings = data.groupby('article_id')['rating'].mean().reset_index()
average_ratings.columns = ['article_id', 'average_rating']
average_ratings
```

## Step 2: Determing the number of ratings (v) for eah article

Step 2: rating_count(v): waxay inoo heesaa inta jeer uu soo laalaabte each id oo kujor columnka article_id


```python
# Step 2: Determine the number of ratings (v) for each article
rating_counts = data['article_id'].value_counts().reset_index()
rating_counts.columns = ['article_id', 'rating_count']
rating_counts
```

##  Step 3: Set the minimum number of ratings required for an article to be on the recommendation list (m = 2)


ratingka ugu yar ee loo ogolyahay articleka si uu ugu soo muuqdo recommendatinka waa 2

```python
# Step 3: Set the minimum number of ratings required for an article to be on the recommendation list (m = 2)
m = 2

```

## Step 4: Calculate the mean rating for all articles (C)


meanka kujira, average ratingka

```python
# Step 4: Calculate the mean rating for all articles (C)
C = average_ratings['average_rating'].mean()
```

## Step 5: Merge average ratings and rating counts


waxaa laysu keenaa: 2dii column: average raing iyo rating count

```python
# Step 5: Merge average ratings and rating counts
merged_data = pd.merge(average_ratings, rating_counts, on='article_id')
merged_data
```

##  Step 6: Apply the updated weighted rating formula

waxaan ku dabaqne formuladii nala siiyo oo ahay:
W = R * v + C * m / v + m

```python
# Step 6: Apply the updated weighted rating formula
merged_data['weighted_rating'] = merged_data.apply(lambda x: ((x['average_rating'] * x['rating_count'] + C * m)) / x['rating_count'] + m, axis=1)
```

```python
merged_data
```

## Step 7: Sort the articles based on weighted rating in descending order

sorting ayaa lagu sameeeye weighted_ratingka oo kan ugu ween ayaa loo kormariye

```python
# Step 7: Sort the articles based on weighted rating in descending order
sorted_articles = merged_data.sort_values('weighted_rating', ascending=False)
```

```python
sorted_articles
```

## Step 8: Select the top 10 articles

soo bandhig 10 ka ugu sareeyo, kadid markaan ku apply gareene qaacidadii weighted rating, sidookalena aan kadhigne sorting oo kala sareen u sameene

```python
# Step 8: Select the top 10 articles
top_10_articles = sorted_articles.head(10)

print(top_10_articles)
```
